```python
#digunakan untuk menentukan kondisi Follower kategori Tinggi
def fuzzyficationFollowerCount_Tinggi(x):
    if(x < 70000):
        return 0
    elif (x >= 80000):
        return 1
    else:
        return (x-70000)/(80000-70000)
        #jika data berada diantara 2 kategori
        
#digunakan untuk menentukan kondisi Follower kategori Sedang
def fuzzyficationFollowerCount_Sedang(x):
    if (40000 <= x < 70000 ):
        return 1
    elif (30000 < x < 40000):
        return (x-30000)/(40000-30000)
        #jika data berada diantara 2 kategori
    elif (70000 < x < 80000):
        return (80000-x)/(80000-70000)
        #jika data berada diantara 2 kategori
    else:
        return 0
#digunakan untuk menentukan kondisi Follower kategori Rendah
def fuzzyficationFollowerCount_Rendah(x):
    if (x <= 30000):
        return 1
    elif (x >= 40000):
        return 0
    else:
        return (40000-x)/(40000-30000)
        #jika data berada diantara 2 kategori
    
#digunakan untuk menentukan kondisi Engagement Rate kategori Rendah
def fuzzyficationEngagementRate_Rendah(x):
    if (x <= 3.0):
        return 1
    elif (x>= 4.0):
        return 0
    else:
        return (4,0-x)/(4.0-3.0)
    #jika data berada diantara 2 kategori
    
#digunakan untuk menentukan kondisi Engagement Rate kategori Sedang
def fuzzyficationEngagementRate_Sedang(x):
    if(4.0 <= x < 7.0):
        return 1
    elif (3.0 < x < 4.0):
        return (x-3.0)/(4.0 - 3.0)
    elif (7.0 < x < 8.0):
        return (8.0-x)/(8.0-7.0)
        #jika data berada diantara 2 kategori
    
#digunakan untuk menentukan kondisi Engagement Rate kategori Tinggi
def fuzzyficationEngagementRate_Tinggi(x):
    if(x < 7.0):
        return 0
    elif(8.0< x <= 10.0):
        return 1
    else :
        return (x - 7.0)/(8.0 - 7.0)
        #jika data berada diantara 2 kategori
    
    
```
